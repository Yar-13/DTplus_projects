В датасете для теста нет полей `closed_at` и `founded_at` , с помощью этих полей мы сможем создать поле `lifetime` в df.

```python
# приведение типов данных
df['founded_at'] = pd.to_datetime(df['founded_at'], format='%Y-%m-%d')
df['first_funding_at'] = pd.to_datetime(df['first_funding_at'], format='%Y-%m-%d')
df['last_funding_at'] = pd.to_datetime(df['last_funding_at'], format='%Y-%m-%d')
# для поля closed_at None означает, что на момент выгрузки даты закрытия стартапа нет,
# пэтому пропуски можно заполнить датой выгрузки, чтобы провести необходимые вычисления.
df['closed_at'] = pd.to_datetime(df['closed_at'].fillna('2018-01-01'))
df['lifetime'] = df['closed_at'] - df['founded_at']
df['lifetime'] = df['lifetime'].dt.days
```

### Категоризация количества ранудов финансирования

```python
def fund_cat(x):
    if x == 1 :
        return '1'
    elif x == 2:
        return '2'
    elif x == 3:
        return '3'
    elif x == 4:
        return '4'
    elif x == 5:
        return '5'
    else:
        return '6+'
```

```python
df_train['round_cat'] = df_train['funding_rounds'].agg(fund_cat)
df_test['round_cat'] = df_test['funding_rounds'].agg(fund_cat)
```

### Категоризация времени жизни стартапа

## Пайплайн и обучение 

```python
RANDOM_STATE = 1337
```

```python
y = df_train['status']
X = df_train.drop(['state_code', 'city', 'status', 'region'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=RANDOM_STATE)
```

```python
ohe_col = ['round_cat', 'more_3_year']
ord_col = ['country_code']
num_col = ['funding_total_usd',
           'funding_rounds',
           'lifetime',
           'usd_per_round',
           'days_funding',
           'first_funding_year',
           'last_funding_year']
```

```python
country_categories = X_train['country_code'].unique().tolist()
```

```python
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',  
            OrdinalEncoder(
                categories=[
                    country_categories
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
) 
```

```python
SS_pipe = Pipeline(
    [
        (
            'simpleImpluter_num',
            SimpleImputer(missing_values=np.nan, strategy='mean')
        ),
        (
            'scaler',
            StandardScaler()
        )
    ]
)

MMS_pipe = Pipeline(
    [
        (
            'simpleImpluter_num',
            SimpleImputer(missing_values=np.nan, strategy='mean')
        ),
        (
            'scaler',
            MinMaxScaler()
        )
    ]
)

RS_pipe = Pipeline(
    [
        (
            'simpleImpluter_num',
            SimpleImputer(missing_values=np.nan, strategy='mean')
        ),
        (
            'scaler',
            RobustScaler()
        )
    ]
)
```

```python
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
)
```

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_col),
        ('ord', ord_pipe, ord_col),
        ('num', SS_pipe, num_col)
    ], 
    remainder='passthrough'
) 
```

```python
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
) 
```

```python
param_grid=[
    {
        'models': [SVC(random_state=RANDOM_STATE, probability=True)],
        'models__kernel': ['sigmoid', 'rbf', 'poly', 'precomputed'],
        'models__degree': range(5, 10),
        'preprocessor__num': [SS_pipe, MMS_pipe, RS_pipe, 'passthrough'],
    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(5,10),
        'models__metric': ['euclidean', 'cityblock'],
        'preprocessor__num': [SS_pipe, MMS_pipe, RS_pipe, 'passthrough'],
    },
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(4, 11),
        'models__min_samples_leaf': range(1, 10), 
        'models__min_samples_split': range(1, 10),
        'preprocessor__num': [SS_pipe, MMS_pipe, RS_pipe, 'passthrough'],
    }
]
```

```python
le = LabelEncoder()

le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
rs = RandomizedSearchCV(
    pipe_final,
    param_grid,
    scoring='f1',
    n_jobs=-1,
    cv=5,
    n_iter=20,
    random_state=RANDOM_STATE
)
```

```python
%%time
rs.fit(X_train, y_train)
```

```python
print(f'Лучшие харктеристики после проведения обучения разных моделей и гиперпараметров:')
display(rs.best_params_)
print(f'Лучшая оенка модели на тренировочныч данных: {rs.best_score_}')
```

```python
y_pred = rs.predict(X_test)
```

```python
f1_score(y_test, y_pred)
```

**Промежуточный вывод:**

Построение пайплайна и обучене моделей дало удовлетворительный результат по обработке данных и получению итоговых метрик на предсказанных данных.
